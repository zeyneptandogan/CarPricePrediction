```python
#Encoding Options (Ref: https://songxia-sophia.medium.com/two-machine-learning-algorithms-to-predict-xgboost-neural-network-with-entity-embedding-caac68717dea)
'''
1. Target encoding: each level of categorical variable is represented by a summary statistic of the target for that level.

2. One-hot encoding: assign 1 to specific category and 0 to other category and transform categorical variable to dummy variable.

3. Label encoding: assign ordinal integer to different categorical levels of categorical variable.

4. Entity embedding: map categorical variables in a function approximation problem into Euclidean spaces, 
and mapping similar values close to each other in the embedding space, and it reveals the intrinsic properties of the categorical variables. 
Usually the mapping is trained by a neural network during the standard supervised training 
process, it not only reduces memory usage and speeds up computational efficiency compared with one-hot encoding. 

'''
```

```python
# define Target & Features & Run Model
def run_model(df,df_test,encode_name,v1,v2,model):
  y = df['price']
  X = df.loc[:,df.columns!='price']


  # Split Train - Test Datasets
  X_train, X_test, y_train ,y_test = train_test_split(X,y,test_size=0.20,random_state=100)

  if model == "xgb":
    # Instantiation
    xgb_r = xgb.XGBRegressor( objective ='reg:linear',
                      n_estimators = 1300, seed = 123, learning_rate=0.50, max_depth=4, colsample_bytree = 1)

    #Hyper-parameter Tuning
    #gs=GridSearchCV(estimator=xgb_r,param_grid=param,scoring='r2',cv=5,verbose=3)
    #gs.fit(X_train,y_train)

    #print(gs.best_params_)
    #print(gs.best_score_)

    
    # Fitting the model
    xgb_r.fit(X_train, y_train)
      
    # Predict the model
    pred = xgb_r.predict(X_test)
      
    # RMSE Computation
    rmse = np.sqrt(MSE(y_test, pred))
    mae = mean_absolute_error(y_test, pred)
    print("Encode: ",encode_name,"| XGB RMSE : % f" %(rmse))
    print("Encode: ",encode_name,"| XGB MAE : % f" %(mae))
    print('R2Score: ',xgb_r.score(X_test,y_test))

    '''  
    ##Create the predicted value csv
    pred_final = pd.DataFrame(pred)

    print(pred_final)

    df_final = df_test.reset_index()

    df_final["price"] = pred_final.reset_index()[0]

    print(df_final)
    
    #df_final["ID"] = df_final[["id"]]

    df_final = df_final[['ID','price']]

    print(df_final.shape)

    df_final.to_csv("submission_tuned_last.csv",index=False)
    
    

    #Feature Importance
    fig,ax = pyplot.subplots(figsize=(v1,v2))
    plot_importance(xgb_r,ax=ax)
    pyplot.show()

    #XGB Decision Tree Visualization

    #fig,ax = pyplot.subplots(figsize=(v1,v2))
    #plot_tree(xgb_r,ax=ax)
    #plt.show()
    '''
    

  elif model == "automl": #https://machinelearningmastery.com/auto-sklearn-for-automated-machine-learning-in-python/
    # define search
    model = AutoSklearnRegressor(time_left_for_this_task=5*60, per_run_time_limit=30, n_jobs=8)
    # perform the search
    model.fit(X_train, y_train )
    # summarize
    print(model.sprint_statistics())
    # evaluate best model
    y_hat = model.predict(X_test)

    rmse = np.sqrt(MSE(y_test, y_hat))
    mae = mean_absolute_error(y_test, y_hat)

    print("Encode:",encode_name," |Auto ML MAE: %.3f" % mae)
    print("Encode:",encode_name," |Auto ML RMSE: %.3f" % rmse)
    print('R2Score: ',model.score(X_test,y_test))

  elif model == "tpot": #http://epistasislab.github.io/tpot/
    # define evaluation procedure
    cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
    # define search
    model = TPOTRegressor(generations=5, population_size=50, scoring='neg_mean_absolute_error', 
                        cv=cv, verbosity=2, random_state=1, n_jobs=-1)
    model.fit(X_train, y_train)
    # export the best model
    model.export('car_price_best_model.py')

    y_hat = model.predict(X_test)

    rmse = np.sqrt(MSE(y_test, y_hat))
    mae = mean_absolute_error(y_test, y_hat)
    print("Encode:",encode_name," |TPOT MAE: %.3f" % mae)
    print("Encode:",encode_name," |TPOT RMSE: %.3f" % rmse)

    


  elif model == "RFT":
    exported_pipeline = RandomForestRegressor(bootstrap=True, max_features=0.1, 
                                              min_samples_leaf=4, min_samples_split=16, n_estimators=100)
    exported_pipeline.fit(X_train, y_train)
    results = exported_pipeline.predict(X_test)

    rmse = np.sqrt(MSE(y_test, results))
    mae = mean_absolute_error(y_test, results)
    print("Encode:",encode_name," |Random Forest MAE: %.3f" % mae)
    print("Encode:",encode_name," |Random Forest RMSE: %.3f" % rmse)
    print('R2Score: ',exported_pipeline.score(X_test,y_test))


  elif model == "DT":
    # Average CV score on the training set was: -1698.8125906693422
    
    exported_pipeline = make_pipeline(
    StackingEstimator(estimator=DecisionTreeRegressor(max_depth=9, min_samples_leaf=5, min_samples_split=20)),
    StackingEstimator(estimator=ElasticNetCV(l1_ratio=0.8500000000000001, tol=0.001)),
    DecisionTreeRegressor(max_depth=8, min_samples_leaf=20, min_samples_split=11))

    # Fix random state for all the steps in exported pipeline
    set_param_recursive(exported_pipeline.steps, 'random_state', 1)

    exported_pipeline.fit(X_train, y_train)
    results = exported_pipeline.predict(X_test)

    rmse = np.sqrt(MSE(y_test, results))
    mae = mean_absolute_error(y_test, results)
    print("Encode:",encode_name," |TPOT DT MAE: %.3f" % mae)
    print("Encode:",encode_name," |TPOT DT Forest RMSE: %.3f" % rmse)
    print('R2Score: ',exported_pipeline.score(X_test,y_test))
```

```python
#Target Encoding

df_train1 = df_train

target_encode_columns = ['brand', 'model', 'transmission', 'fuelType']
target = ['price']
target_encode_df = df_train1[target_encode_columns + target].reset_index().drop(columns = 'ID', axis = 1)

target_name = target[0]
target_df = pd.DataFrame()

for embed_col in target_encode_columns:
    val_map = target_encode_df.groupby(embed_col)[target].mean().to_dict()[target_name]
    target_df[embed_col] = target_encode_df[embed_col].map(val_map).values
    
score_target_drop = df_train1.drop(target_encode_columns, axis = 1).reset_index().drop(columns = 'ID', axis = 1)
score_target = pd.concat([score_target_drop, target_df], axis = 1)




#print(score_target)

run_model(score_target,score_target,"Target Encoding",10,8,"xgb")
#run_model(score_target,"Target Encoding",10,8,"automl")
#run_model(score_target,"Target Encoding",10,8,"tpot")
#run_model(score_target,"Target Encoding",10,8,"RFT")
#run_model(score_target,score_target,"Target Encoding",10,8,"DT")
```

```python
#Label Encoding

df_train2 = df_train

encode_columns = ['brand', 'model', 'transmission', 'fuelType']


encode_df = df_train2[encode_columns]
encode_df = encode_df.astype('str')
encode_df = encode_df.apply(LabelEncoder().fit_transform)
score_encode_drop = df_train2.drop(encode_columns, axis = 1)
score_encode = pd.concat([score_encode_drop, encode_df], axis = 1)

df_test2 = df_test

encode_columns = ['brand', 'model', 'transmission', 'fuelType']


encode_df = df_test2[encode_columns]
encode_df = encode_df.astype('str')
encode_df = encode_df.apply(LabelEncoder().fit_transform)
score_encode_drop = df_test2.drop(encode_columns, axis = 1)
score_encode_test = pd.concat([score_encode_drop, encode_df], axis = 1)

#score_encode

run_model(score_encode,score_encode_test,"Label Encoding",30,30,"xgb")
#run_model(score_encode,"Label Encoding",10,8,"automl")
#run_model(score_encode,"Label Encoding",10,8,"tpot")
#run_model(score_encode,"Label Encoding",10,8,"RFT")
#run_model(score_target,score_target,"Label Encoding",10,8,"DT")
#tuning(score_encode)
```

```python
#One-Hot Encoding

df_train3 = df_train
df_train3_test = df_test

onehot_columns = ['brand', 'model', 'transmission', 'fuelType']


onehot_df = df_train3[onehot_columns]
onehot_df = pd.get_dummies(onehot_df, columns = onehot_columns)
score_onehot_drop = df_train3.drop(onehot_columns, axis = 1)
score_onehot = pd.concat([score_onehot_drop, onehot_df], axis = 1)



onehot_df_test = df_train3_test[onehot_columns]
onehot_df_test = pd.get_dummies(onehot_df_test, columns = onehot_columns)
score_onehot_drop_test = df_train3_test.drop(onehot_columns, axis = 1)
score_onehot_test = pd.concat([score_onehot_drop_test, onehot_df_test], axis = 1)


train_cols = list(score_onehot.columns)
test_cols = list(score_onehot_test.columns)
cols_not_in_test = {c:0 for c in train_cols if c not in test_cols}
score_onehot_test = score_onehot_test.assign(**cols_not_in_test)

a = score_onehot.columns.intersection(score_onehot_test.columns)
score_onehot_test = score_onehot_test[a]

score_onehot_test.drop('price',inplace=True,axis=1)


print(score_onehot.shape)
print(score_onehot_test.shape)

#score_onehot

run_model(score_onehot,score_onehot_test,"One-Hot Encoding",30,35,"xgb")
#run_model(score_onehot,score_onehot_test,"One-Hot Encoding",10,8,"automl")
#run_model(score_encode,"One-Hot Encoding",10,8,"tpot")
#run_model(score_onehot,score_onehot_test,"One-Hot Encoding",20,25,"RFT")
#run_model(score_onehot,score_onehot_test,"One-Hot Encoding",20,25,"DT")
#tuning(score_onehot)
```

```python
from sklearn.model_selection import GridSearchCV

def tuning (df):

  y = df['price']
  X = df.loc[:,df.columns!='price']

  # Split Train - Test Datasets
  X_train, X_test, y_train ,y_test = train_test_split(X,y,test_size=0.2,random_state=100)

  params = { 'max_depth': [3,6,10],
            'learning_rate': [0.01, 0.05, 0.1],
            'n_estimators': [100, 500, 1000],
            'colsample_bytree': [0.3, 0.7]}

  xgbr = xgb.XGBRegressor(seed = 20)
  clf = GridSearchCV(estimator=xgbr, 
                    param_grid=params,
                    scoring='neg_mean_squared_error', 
                    verbose=3)
  clf.fit(X_train, y_train)
  print("Best parameters:", clf.best_params_)
  print("Lowest RMSE: ", (-clf.best_score_)**(1/2.0))
```
